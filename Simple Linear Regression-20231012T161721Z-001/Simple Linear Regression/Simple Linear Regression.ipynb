```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing dataset
```

```python
dataset = pd.read_csv('/content/drive/My Drive/Process mining project/Simple Linear Regression/Salary_Data.csv')
X_Axis = dataset.iloc[:, :-1].values
Y_Axis = dataset.iloc[:, -1].values
```

```python
#Splitting Dataset into the training set and test set
```

```python
from sklearn.model_selection import train_test_split
X_training, X_test, y_training, y_test = train_test_split(X_Axis, Y_Axis, test_size = 1/3, random_state = 0)
```

```python
#Training the Simple Linear Regression Model on the Training Set
```

```python
from sklearn.linear_model import LinearRegression
regressor_for_data = LinearRegression()
regressor_for_data.fit(X_training, y_training)
```

```python
#Predicting Test Results
```

```python
y_predictor = regressor_for_data.predict(X_test)
```

```python
#Visualizing the Training Set Results
```

```python
plt.scatter(X_training, y_training, color = 'purple')
plt.plot(X_training, regressor_for_data.predict(X_training), color = 'red')
plt.title('Income vs Experience (Training set)')
plt.xlabel('# of Experience')
plt.ylabel('Income')
plt.show()
```

```python
#Visualizing the Test Set Results
```

```python
plt.scatter(X_test, y_test, color = 'orange')
plt.plot(X_training, regressor_for_data.predict(X_training), color = 'purple')
plt.title('Income vs Experience(Test set)')
plt.xlabel('# of Experience')
plt.ylabel('Income')
plt.show()
```
